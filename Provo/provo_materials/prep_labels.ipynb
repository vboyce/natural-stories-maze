```python
# get freqs from wordfreq
import csv
from wordfreq import zipf_frequency
with open('words.csv') as csvfile:
    f = csv.reader(csvfile)
    next(f) #Deal with header row
    with open('freq.csv', 'w') as outfile:
        out=csv.writer(outfile, delimiter=",")
        for row in f:
            word=row[4]
            freq=zipf_frequency(word, "en")
            print(word)
            print(freq)
            out.writerow([row[0],row[1],row[2],word,freq])
```

```python
from nltk.tokenize import word_tokenize
import csv
with open('sents.csv') as csvfile:
    f = csv.reader(csvfile)
    next(f)
    with open('sent.txt', 'w') as outf:
        for row in f:
            sent=row[2]
            tokens=' '.join(word_tokenize(sent))+" <eos>\n"
            outf.write(tokens)
```

```python
from nltk.tokenize import word_tokenize
import csv
with open('words.csv') as csvfile:
    f = csv.reader(csvfile)
    next(f)
    with open('tokens.csv', 'w') as outf:
        out=csv.writer(outf, delimiter=",")
        for row in f:
            word=row[4]
            tokens=word_tokenize(word)
            word_list=[word,]
            word_list.extend(tokens)
            out.writerow(word_list)
```
